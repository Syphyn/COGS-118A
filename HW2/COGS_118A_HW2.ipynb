$f(x) = \lambda(1-x^TAx)$

$\frac{\partial f(x)}{\partial x} = \lambda \frac{\partial}{\partial x} (1-x^TAx) = -2\lambda Ax$

## 7 (10 points) Ploting Decision Boundaries {-}

#### 7.1 {-}
$w_1x_1 + w_2x_2 + b = 0$ where $(w_1, w_2, b) = (1, 1, -8)$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%config InlineBackend.figure_format ='retina'

def plot_line(w1, w2, b, num = 30):
    X_range = np.linspace(0,10,num)
    X1_line = X_range
    X2_line = (b + w1 * X1_line) / (-w2)
    
    fig = plt.figure(figsize = (6, 6))
    plt.plot(X1_line, X2_line)
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.grid(linestyle='--')
    plt.title("Line: ${}x_1+{}x_2+{}=0$".format(w1,w2,b))
    plt.show()

plot_line(1, 1, -8)
```

#### 7.2 {-}
$w_1x_1 + w_2x_2 + w_3x_3 + b = 0$ where $(w_1, w_2, w_3, b) = (3, 5, 2, -8)$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%config InlineBackend.figure_format ='retina'

def plot_plane(w1, w2, w3, b, num = 30):
  '''Helper function for 7.2; modify this example using plt.plot()instead of Axes3D() to handle 7.1 & 7.3''' 

  X_range = np.linspace(0,10,num)

  # sets up space to plot, 3d equivalent of np.linspace() used for 2d plotting
  X1_plane, X2_plane, X3_plane = np.meshgrid(X_range, X_range, X_range)

  # this defines the equation to plot
  X3_plane = (b + w1 * X1_plane + w2 * X2_plane)/(-w3)

  # does plotting
  fig = plt.figure(figsize = (6, 6))
  ax = Axes3D(fig, elev = -150, azim = 130)
  ax.scatter(X1_plane, X2_plane, X3_plane)
  ax.set_xlabel('$x_1$')
  ax.set_ylabel('$x_2$')
  ax.set_zlabel('$x_3$')
  plt.title("Plane: ${}x_1+{}x_2+{}x_3+{}=0$".format(w1,w2,w3,b))
  plt.show()

plot_plane(3, 5, 2, -8)
```

#### 7.3 {-}
$w_1x_1^2 + w_2x_2^2 + b = 0$ where $(w_1, w_2, b) = (1, 1, -10)$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%config InlineBackend.figure_format ='retina'

def plot_cicle(w1, w2, b, num = 30):
    X1_range = np.linspace(-5,5,num)
    X2_range = np.linspace(-5,5,num)
    X1, X2 = np.meshgrid(X1_range,X2_range)
    F = w1*(X1**2) + w2*(X2**2) + b
    
    fig = plt.figure(figsize = (6, 6))
    plt.contour(X1, X2, F, [0])
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.grid(linestyle='--')
    plt.title("Circle: ${}x_1+{}x_2+{}=0$".format(w1,w2,b))
    plt.show()

plot_cicle(1, 1, -10)
```
